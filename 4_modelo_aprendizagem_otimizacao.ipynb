### Leitura dos Dados

```python
df_ini  = pd.read_csv("data/dados.csv")
df_prep = pd.read_csv("data/dados_prep.csv")
df_ind  = pd.read_csv("data/dados_prep_index.csv")
```

### Modelo de Predição para Order_Priority para o valor 'Not Specified''

```python
df_esp     = df_ind.query(''' Order_Priority !='Not Specified' ''')
```

```python
df_esp[['Order_Priority']].sample(5)
```

```python
def order_pri(line):
    if line == 'Critical':
        return 1   
    elif line == 'High':
        return 2   
    elif line == 'Low':
        return 3   
    elif line == 'Medium':
        return 4   
```

```python
df_esp['Type_Order_Priority'] = df_esp['Order_Priority'].apply(lambda x: order_pri(x))
```

```python
df_esp.sample(5)
```

```python
df_not_spec = df_ind.query(''' Order_Priority =='Not Specified' ''')
```

```python
df_esp.columns
```

```python
df = df_esp
```

```python
X =  df.drop(['Order_ID','Profit', 'Order_Priority','Type_Order_Priority', ], axis = 1)
Y =  df['Type_Order_Priority']
```

```python
seed = 1234
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state = seed )
```

```python
KNN_model = KNeighborsClassifier(n_neighbors=5)
```

```python
KNN_model.fit(X_train, y_train)
KNN_prediction = KNN_model.predict(X_test)
```

```python
print(accuracy_score(KNN_prediction, y_test))
```

```python
print(classification_report(KNN_prediction, y_test))
```

### Execução com N-Modelos 

```python
models = []

models.append(("LogisticRegression",LogisticRegression()))
models.append(("SVC",SVC()))
models.append(("LinearSVC",LinearSVC()))
models.append(("KNeighbors",KNeighborsClassifier()))
models.append(("DecisionTree",DecisionTreeClassifier()))
models.append(("RandomForest",RandomForestClassifier()))
rf2 = RandomForestClassifier(n_estimators=100, criterion='gini',
                                max_depth=10, random_state=0, max_features=None)
models.append(("RandomForest2",rf2))
models.append(("MLPClassifier",MLPClassifier(solver='lbfgs', random_state=0)))


results = []
names = []
for name,model in models:
    result = cross_val_score(model, X,Y,  cv=3)
    names.append(name)
    results.append(result)

for i in range(len(names)):
    print(names[i],results[i].mean())
```
